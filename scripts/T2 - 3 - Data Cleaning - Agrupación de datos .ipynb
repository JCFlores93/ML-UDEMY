## Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500
gender_data = []
income_data = []

for i in range(0, 500): 
    #Para elegir data aleatoria de gender(array)
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data
```

```python
#N(m, s) -> m(media) + s(desviacion tipíca) * Z (normal)
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income  = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        'Gender': gender_data,
        "Economic Status": income_data,
        'Height': height,
        'Weight': weight,
        'Age': age,
        'Income': income
    }
)
```

```python
data.head()
```

## Agrupacion de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for name, groups in grouped_gender:
    print(name)
    print(groups)
```

```python
grouped_gender.get_group('Female')
```

```python
#Cuando deseamos obtener la data de 2 columnas
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
##Normal
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
#Deberimos leer variable por variable 
grouped_income = double_group["Income"]
```

```python
grouped_income.describe()
```

```python
## agregated agregar conjuntamente 
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    }
)
```

```python
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h: np.mean(h)/np.std(h)
    }
)
```

```python
## Con aggregate las funciones se van a agregar a cada una de las variables
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x)])
```
